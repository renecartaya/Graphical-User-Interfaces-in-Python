```python
from tkinter import *
import operator

root = Tk()
root.title('Figures draw test GUI v1.0')
root.iconbitmap('C:/Users/ASUS/Pictures/Logos/generic-logo.ico')
root.geometry("500x500")

canvas = Canvas(root, width=400, height=400)
canvas.pack(side=LEFT, padx=10, pady=10)

# Square coordinates
corner_a_1 = (10, 25)
corner_a_2 = (200, 25)
corner_a_3 = (10, 185)
corner_a_4 = (200, 185)

# Draw the square
square = canvas.create_rectangle(corner_a_1[0], corner_a_1[1], corner_a_4[0], corner_a_4[1], fill='blue')

# Slider to control the square's vertical position
slider = Scale(root, from_=0, to=200, orient=HORIZONTAL, length=400)
slider.place(x = 70, y = 10)

# Function to update the square's position based on the slider value
def update_square_position(value):
    canvas.move(square, 0, int(value) - corner_a_1[1])

# Link the slider to the update function
slider.config(command=update_square_position)

root.mainloop()
```

```python
from tkinter import *

root = Tk()
root.title('Figures draw test GUI v1.0')
root.iconbitmap('C:/Users/ASUS/Pictures/Logos/generic-logo.ico')
root.geometry("500x500")

canvas = Canvas(root, width=400, height=400)
canvas.pack(side=LEFT, padx=10, pady=10)

# Square coordinates
corner_a_1 = (10, 25)
corner_a_2 = (200, 25)
corner_a_3 = (10, 185)
corner_a_4 = (200, 185)

# Draw the square
square = canvas.create_rectangle(corner_a_1[0], corner_a_1[1], corner_a_4[0], corner_a_4[1], fill='blue')

# Slider to control the square's vertical position
slider = Scale(root, from_=0, to=200, orient=HORIZONTAL, length=400)
slider.place(x = 70, y = 10)

# Function to update the square's position based on the slider value
def update_square_position(value):
    y = int(value)
    canvas.coords(square, corner_a_1[0], y, corner_a_4[0], y + (corner_a_4[1] - corner_a_4[0]))

# Link the slider to the update function
slider.config(command=update_square_position)

root.mainloop()
```

```python
from tkinter import *
import operator

root = Tk()
root.title('Figures draw test GUI v1.0')
root.iconbitmap('C:/Users/ASUS/Pictures/Logos/generic-logo.ico')
root.geometry("500x500")

canvas = Canvas(root, width=400, height=400)
canvas.pack(side=LEFT, padx=10, pady=10)

# Square coordinates
corner_a_1 = (10, 25)
corner_a_2 = (200, 25)
corner_a_3 = (10, 185)
corner_a_4 = (200, 185)

# Draw the square
square = canvas.create_rectangle(corner_a_1[0], corner_a_1[1], corner_a_4[0], corner_a_4[1], fill='blue')

# Slider to control the square's vertical position
slider_y = Scale(root, from_=0, to=200, orient=HORIZONTAL, length=400)
slider_y.place(x=70, y=10)

# Function to update the square's position based on the slider value
def update_square_position_y(value):
    canvas.move(square, 0, int(value) - corner_a_1[1])

# Link the slider to the update function
slider_y.config(command=update_square_position_y)

# Slider to control the square's horizontal position
slider_x = Scale(root, from_=0, to=190, orient=VERTICAL, length=200)
slider_x.place(x=10, y=220)

# Function to update the square's position based on the slider value
def update_square_position_x(value):
    canvas.move(square, int(value) - corner_a_1[0], 0)

# Link the slider to the update function
slider_x.config(command=update_square_position_x)

root.mainloop()
```
