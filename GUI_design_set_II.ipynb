```python
###########################################################################################################################
#
#                                                 Test of Figures Drawing in Tkinter
#                  
###########################################################################################################################

from tkinter import *

root = Tk()
root.title('Figures draw test GUI v1.0')
root.iconbitmap('C:/Users/ASUS/Pictures/Logos/generic-logo.ico')
root.geometry("500x500")

canvas = Canvas()

canvas.create_line(15, 25, 200, 25, width=5)
canvas.pack()

canvas.create_line(300, 35, 300, 200, dash=(10), width=5) # Dashed line
canvas.pack()

canvas.create_line(55, 85, 155, 85, 105, 180, 55, 85, width=5) # Triangle
canvas.pack()

canvas.create_oval(150, 150, 180, 180, outline = "black", fill = "white",width = 2) # Small oval
canvas.pack()
 
root.mainloop()
```

```python
###########################################################################################################################
#
#                                                 Test of Figures Drawing in Tkinter
#                  
###########################################################################################################################

from tkinter import *

root = Tk()
root.title('Figures draw test GUI v1.0')
root.iconbitmap('C:/Users/ASUS/Pictures/Logos/generic-logo.ico')
root.geometry("500x500")

canvas = Canvas()

#points for the vertices in x1,y1,x2,y2, and so on
 
points = [150, 100, 200, 120, 240, 180, 210, 200, 150, 150, 100, 200]
 
canvas.create_polygon(points, outline = "blue", fill = "orange", width = 2)
canvas.pack()

canvas.create_arc(30,200,90,100,extent =210,outline ="black",fill ="white", width =2)
canvas.pack()
 
root.mainloop()
```
