```python
###########################################################################################################
#
#                                      Gauge simple  
#
###########################################################################################################

import tkinter as tk
from math import pi, cos, sin

root = tk.Tk()
root.title('GUI Slider Test')
root.configure(background = '#e0e0e0')
root.iconbitmap('C:/Users/ASUS/Pictures/symbol_light_S.ico')
root.geometry("260x180") # set the window size
#root.columnconfigure(0, weight=1)
#root.columnconfigure(1, weight=3)

class Meter(tk.Frame):
    def __init__(self, master=None, **kw):
        tk.Frame.__init__(self, master, **kw)

        self.var = tk.IntVar(self, 0)

        self.canvas = tk.Canvas(self, 
                                width=200, 
                                height=110,
                                borderwidth=2, 
                                relief='sunken',
                                bg='white')
        
        self.scale = tk.Scale(self, 
                              orient='horizontal', 
                              from_=0, 
                              to=100, 
                              variable=self.var)

        self.meter = self.canvas.create_line(100, 100, 10, 100,
                                             fill='black',
                                             width=3,
                                             arrow='last')
        self.angle = 0
        self.updateMeterLine(0.2)

        self.canvas.create_arc(10, 10, 190, 190, 
                               extent=108, 
                               start=36,
                               style='arc', 
                               outline='red')

        self.canvas.pack(fill='both')
        self.scale.pack()

        self.var.trace_add('write', self.updateMeter)  # if this line raises an error, change it to the old way of adding a trace: self.var.trace('w', self.updateMeter)

    def updateMeterLine(self, a):
        """Draw a meter line"""
        self.angle = a
        x = 100 - 90 * cos(a * pi)
        y = 100 - 90 * sin(a * pi)
        self.canvas.coords(self.meter, 100, 100, x, y)

    def updateMeter(self, name1, name2, op):
        """Convert variable to angle on trace"""
        mini = self.scale.cget('from')
        maxi = self.scale.cget('to')
        pos = (self.var.get() - mini) / (maxi - mini)
        self.updateMeterLine(pos * 0.6 + 0.2)

if __name__ == '__main__':
#    root = tk.Tk()
    meter = Meter(root)
    meter.pack()
    root.mainloop()
```

```python
###########################################################################################################
#
#                                      Gauge with vanish effect   
#
###########################################################################################################

import tkinter as tk
from math import pi, cos, sin

root = tk.Tk()
root.title('GUI Slider Test')
root.configure(background = '#e0e0e0')
root.iconbitmap('C:/Users/ASUS/Pictures/symbol_light_S.ico')
root.geometry("260x180") # set the window size
#root.columnconfigure(0, weight=1)
#root.columnconfigure(1, weight=3)


class Meter(tk.Frame):
    def __init__(self, master=None, **kw):
        tk.Frame.__init__(self, master, **kw)

        self.meter = []
        self.angle = []
        self.var = tk.IntVar(self, 0)

        self.canvas = tk.Canvas(self, 
                                width=200, 
                                height=110,
                                borderwidth=2, 
                                relief='sunken',
                                bg='white')
        
        self.scale = tk.Scale(self, 
                              orient='horizontal', 
                              from_=0, 
                              to=100, 
                              variable=self.var)

        for j, i in enumerate(range(0, 100, 5)):
            self.meter.append(self.canvas.create_line(100, 100, 10, 100,
                                                      fill='grey%i' % i,
                                                      width=3,
                                                      arrow='last'))
            self.angle.append(0)
            self.canvas.lower(self.meter[j])
            self.updateMeterLine(0.2, j)

        self.canvas.create_arc(10, 10, 190, 190, extent=108, start=36,
                               style='arc', outline='red')

        self.canvas.pack(fill='both')
        self.scale.pack()

        self.var.trace_add('write', self.updateMeter)  # if this line raises an error, change it to the old way of adding a trace: self.var.trace('w', self.updateMeter)
        self.updateMeterTimer()

    def updateMeterLine(self, a, l=0):
        """Draw a meter line (and recurse for lighter ones...)"""
        oldangle = self.angle[l]
        self.angle[l] = a
        x = 100 - 90 * cos(a * pi)
        y = 100 - 90 * sin(a * pi)
        self.canvas.coords(self.meter[l], 100, 100, x, y)
        l += 1
        if l < len(self.meter):
            self.updateMeterLine(oldangle, l)

    def updateMeter(self, name1, name2, op):
        """Convert variable to angle on trace"""
        mini = self.scale.cget('from')
        maxi = self.scale.cget('to')
        pos = (self.var.get() - mini) / (maxi - mini)
        self.updateMeterLine(pos * 0.6 + 0.2)

    def updateMeterTimer(self):
        """Fade over time"""
        self.var.set(self.var.get())
        self.after(20, self.updateMeterTimer)


if __name__ == '__main__':
#    root = tk.Tk()
    meter = Meter(root)
    meter.pack()
    root.mainloop()
```

```python
import tkinter as tk
from tkinter import ttk
from PIL import Image, ImageTk, ImageDraw
from ttkbootstrap import Style


class RadialGauge(ttk.Label):
    def __init__(self, parent, styler=None, style='primary.TLabel', **kwargs):

        excluded = [
            'anchor', 'theme', 'image', 'compound', 'relief', 'bordwidth',
            'maximum', 'value', 'variable', 'minimum',
            'indicatorcolor', 'troughcolor', 'textvariable', 'size'
        ]

        kwargs_ = {k: v for k, v in kwargs.items() if k not in excluded}
        super().__init__(parent, class_='RadialGauge', style=style, **kwargs_)

        self.styler = styler or Style()

        self.indicatorcolor = kwargs.get(
            'indicatorcolor'
        ) or self.styler.lookup(style, 'foreground') or '#1f77b4'

        self.troughcolor = kwargs.get(
            'troughcolor'
        ) or self.styler.lookup('Horizontal.TProgressbar', 'troughcolor') or '#cccccc'

        self.maximum = max(kwargs.get('maximum', 100), 1)
        self.minimum = kwargs.get('minimum', 0)

        self.textvariable = kwargs.get(
            'textvariable',
            tk.StringVar(value=kwargs.get('text', ''))
        )
        self.configure(textvariable=self.textvariable)

        self.variable = kwargs.get(
            'variable',
            tk.IntVar(value=kwargs.get('value', 0))
        )
        self.variable.trace_add('write', self.update_indicator)

        self.size = kwargs.get('size', 120)

        self.im = None
        self.arc = None

        self.configure(compound='center', anchor='center')
        self.bind('<Configure>', self.resize)

        self.update_indicator()

    def resize(self, event):
        padding = 6
        self.size = max(
            min(event.width - padding, event.height - padding),
            80
        )
        self.update_indicator()

    def step(self, delta=1):
        value = self.variable.get()
        if value + delta <= self.maximum:
            self.variable.set(value + delta)

    def down(self, delta=1):
        value = self.variable.get()
        if value - delta >= self.minimum:
            self.variable.set(value - delta)

    def update_indicator(self, *args):
        value = max(self.minimum, min(self.variable.get(), self.maximum))
        self.variable.set(value)

        ratio = value / self.maximum

        start_angle = 135
        extent = int(ratio * 270)

        self.im = Image.new('RGBA', (1000, 1000), (0, 0, 0, 0))
        draw = ImageDraw.Draw(self.im)

        draw.arc(
            (0, 0, 990, 990),
            start_angle,
            start_angle + 270,
            fill=self.troughcolor,
            width=100
        )

        draw.arc(
            (0, 0, 990, 990),
            start_angle,
            start_angle + extent,
            fill=self.indicatorcolor,
            width=100
        )

        self.arc = ImageTk.PhotoImage(
            self.im.resize((self.size, self.size), Image.BICUBIC)
        )

        self.configure(image=self.arc)


if __name__ == '__main__':
    root = tk.Tk()
    root.geometry('400x400')

    try:
        root.iconbitmap('C:/Users/ASUS/Pictures/symbol_light_S.ico')
    except tk.TclError:
        pass  # icon is optional

    style = Style('cosmo')
    root.title('Dial Knob Test')
    root.configure(background='#e0e0e0')

    gauge = RadialGauge(
        root,
        styler=style,
        font='Helvetica 14 bold',
        maximum=360,
        value=10,
        text='10 rpm'
    )

    gauge.pack(fill='both', expand=True, padx=10, pady=20)

    gauge.variable.trace_add(
        'write',
        lambda *a, g=gauge: g.textvariable.set(f'{g.variable.get()} rpm')
    )

    ttk.Scale(
        root,
        from_=0,
        to=360,
        variable=gauge.variable
    ).pack(fill='x', padx=10, pady=10)

    ttk.Button(
        root,
        text='Increase RPM',
        command=gauge.step
    ).pack(fill='x', padx=10, pady=5)

    ttk.Button(
        root,
        text='Decrease RPM',
        command=gauge.down
    ).pack(fill='x', padx=10, pady=5)

    root.mainloop()
```

```python
import tkinter as tk

# ------------------ CONSTANTS ------------------
WIDTH, HEIGHT = 1100, 600

BG_COLOR = "#1e1e2e"
PANEL_COLOR = "#0f172a"
BTN_COLOR = "#24314f"

ACCENT1 = "#3b82f6"
ACCENT2 = "#facc15"


# ------------------ MAIN APP ------------------
class DashboardApp(tk.Tk):
    def __init__(self):
        super().__init__()

        self.title("Car Dashboard")
        self.geometry(f"{WIDTH}x{HEIGHT}")
        self.resizable(False, False)
        self.configure(bg=BG_COLOR)

        # App icon (safe)
        try:
            self.iconbitmap("C:/Users/ASUS/Pictures/symbol_light_S.ico")
        except tk.TclError:
            pass

        self.create_navbar()
        self.create_dashboard()

    # ------------------ NAVBAR ------------------
    def create_navbar(self):
        nav = tk.Frame(self, bg=BG_COLOR)
        nav.pack(pady=10)

        for name in ("DASHBOARD", "AC", "MUSIC", "MAP"):
            tk.Button(
                nav,
                text=name,
                width=12,
                height=2,
                bg=BTN_COLOR,
                fg="white",
                relief="flat",
                font=("Segoe UI", 10, "bold")
            ).pack(side="left", padx=10)

    # ------------------ DASHBOARD ------------------
    def create_dashboard(self):
        canvas = tk.Canvas(self, bg=BG_COLOR, highlightthickness=0)
        canvas.pack(fill="both", expand=True)

        # Gauges
        self.draw_gauge(canvas, 250, 260, 100, 20, 40, ACCENT2, "Indoor Temp")
        self.draw_gauge(canvas, 850, 260, 100, 29, 40, ACCENT1, "Outdoor Temp")

        # Weather panel
        panel = tk.Frame(self, bg=PANEL_COLOR, width=300, height=300)
        panel.place(x=400, y=150)

        tk.Label(
            panel,
            text="Weather Forecast",
            bg=PANEL_COLOR,
            fg="#fbbf24",
            font=("Segoe UI", 14, "bold")
        ).pack(pady=10)

        tk.Label(
            panel,
            text="‚òÅ Cloudy",
            bg=PANEL_COLOR,
            fg="white",
            font=("Segoe UI", 16)
        ).pack(pady=10)

        tk.Label(
            panel,
            text="Precipitation: 20%\nHumidity: 70%\nWind: 32 km/h",
            bg=PANEL_COLOR,
            fg="white",
            font=("Segoe UI", 10)
        ).pack(pady=10)

        modes = tk.Frame(panel, bg=PANEL_COLOR)
        modes.pack(pady=20)

        for m in ("‚ùÑ Mode1", "üåÄ Mode2", "üî• Mode3"):
            tk.Label(
                modes,
                text=m,
                bg=PANEL_COLOR,
                fg="white",
                font=("Segoe UI", 10)
            ).pack(side="left", padx=15)

    # ------------------ GAUGE DRAW ------------------
    def draw_gauge(self, canvas, x, y, r, value, max_value, color, label):
        canvas.create_oval(
            x - r, y - r, x + r, y + r,
            outline="#2a2a3a",
            width=15
        )

        extent = (value / max_value) * 270
        canvas.create_arc(
            x - r, y - r, x + r, y + r,
            start=135,
            extent=extent,
            style="arc",
            outline=color,
            width=15
        )

        canvas.create_text(
            x, y - 10,
            text=f"{value}¬∞C",
            fill=color,
            font=("Segoe UI", 26, "bold")
        )

        canvas.create_text(
            x, y + 30,
            text=label,
            fill="white",
            font=("Segoe UI", 12)
        )


# ------------------ RUN APP ------------------
if __name__ == "__main__":
    app = DashboardApp()
    app.mainloop()
```
