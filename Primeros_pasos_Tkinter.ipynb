```python
"""
Ejemplo 1B: Versi√≥n mejorada con m√°s widgets
"""
import tkinter as tk

class HolaMundoApp:
    def __init__(self, ventana):
        self.ventana = ventana
        self.ventana.title("Hola Mundo - Mejorado")
        self.ventana.geometry("400x300")
        
        # Contador de clicks
        self.contador_clicks = 0
        
        self.crear_widgets()
    
    def crear_widgets(self):
        # T√≠tulo
        tk.Label(
            self.ventana,
            text="EJEMPLO DE GUI INTERACTIVA",
            font=("Arial", 16, "bold"),
            fg="darkblue"
        ).pack(pady=10)
        
        # L√≠nea separadora
        tk.Frame(self.ventana, height=2, bg="gray").pack(fill="x", padx=20, pady=10)
        
        # Etiqueta principal
        self.etiqueta_principal = tk.Label(
            self.ventana,
            text="Presiona el bot√≥n para cambiar el mensaje",
            font=("Arial", 12),
            fg="green"
        )
        self.etiqueta_principal.pack(pady=10)
        
        # Bot√≥n de acci√≥n
        self.boton_saludo = tk.Button(
            self.ventana,
            text="üëâ Haz Click Aqu√≠ üëà",
            command=self.cambiar_mensaje,
            bg="#4CAF50",  # Verde
            fg="white",
            font=("Arial", 12, "bold"),
            width=20,
            height=2,
            relief="raised",
            cursor="hand2"  # Cambia el cursor al pasar por encima
        )
        self.boton_saludo.pack(pady=20)
        
        # Contador de clicks
        self.etiqueta_contador = tk.Label(
            self.ventana,
            text=f"Clicks: {self.contador_clicks}",
            font=("Arial", 10),
            fg="gray"
        )
        self.etiqueta_contador.pack()
        
        # Bot√≥n para resetear
        tk.Button(
            self.ventana,
            text="Reiniciar",
            command=self.reiniciar,
            bg="lightgray",
            font=("Arial", 10)
        ).pack(pady=10)
    
    def cambiar_mensaje(self):
        # Incrementar contador
        self.contador_clicks += 1
        
        # Cambiar mensaje basado en n√∫mero de clicks
        mensajes = [
            "¬°Hola Mundo!",
            "¬°Bienvenido a Mecatr√≥nica!",
            "¬°Las GUI son poderosas!",
            "¬°Sigue aprendiendo!",
            "¬°Eres incre√≠ble!"
        ]
        
        # Ciclar entre mensajes
        mensaje_index = (self.contador_clicks - 1) % len(mensajes)
        
        # Actualizar etiqueta
        self.etiqueta_principal.config(
            text=mensajes[mensaje_index],
            fg=["red", "blue", "green", "purple", "orange"][mensaje_index]
        )
        
        # Actualizar contador
        self.etiqueta_contador.config(text=f"Clicks: {self.contador_clicks}")
        
        # Imprimir en consola
        print(f"Click #{self.contador_clicks}: {mensajes[mensaje_index]}")
    
    def reiniciar(self):
        self.contador_clicks = 0
        self.etiqueta_principal.config(
            text="Presiona el bot√≥n para cambiar el mensaje",
            fg="green"
        )
        self.etiqueta_contador.config(text="Clicks: 0")
        print("Contador reiniciado")

# Ejecutar la aplicaci√≥n
if __name__ == "__main__":
    root = tk.Tk()
    app = HolaMundoApp(root)
    root.mainloop()
```

```python
"""
Calculadora SIMPLIFICADA para explicaci√≥n en clase
"""
import tkinter as tk

# 1. Crear ventana
ventana = tk.Tk()
ventana.title("Calculadora Simple")
ventana.geometry("300x250")

# 2. Crear entradas
tk.Label(ventana, text="N√∫mero 1:", font=("Arial", 11)).pack(pady=5)
entrada1 = tk.Entry(ventana, font=("Arial", 12), width=15)
entrada1.pack(pady=5)

tk.Label(ventana, text="N√∫mero 2:", font=("Arial", 11)).pack(pady=5)
entrada2 = tk.Entry(ventana, font=("Arial", 12), width=15)
entrada2.pack(pady=5)

# 3. Etiqueta para resultado
resultado_label = tk.Label(
    ventana,
    text="Resultado: --",
    font=("Arial", 12, "bold"),
    fg="blue"
)
resultado_label.pack(pady=10)

# 4. Funci√≥n para calcular
def calcular(operacion):
    try:
        num1 = float(entrada1.get())
        num2 = float(entrada2.get())
        
        if operacion == "+":
            resultado = num1 + num2
            simbolo = "+"
        elif operacion == "-":
            resultado = num1 - num2
            simbolo = "-"
        elif operacion == "*":
            resultado = num1 * num2
            simbolo = "√ó"
        elif operacion == "/":
            if num2 == 0:
                resultado_label.config(text="Error: Divisi√≥n por 0", fg="red")
                return
            resultado = num1 / num2
            simbolo = "√∑"
        
        resultado_label.config(
            text=f"Resultado: {num1} {simbolo} {num2} = {resultado}",
            fg="green"
        )
        print(f"C√°lculo: {num1} {operacion} {num2} = {resultado}")
        
    except ValueError:
        resultado_label.config(text="Error: Ingresa n√∫meros", fg="red")

# 5. Botones de operaciones
frame_botones = tk.Frame(ventana)
frame_botones.pack(pady=10)

tk.Button(frame_botones, text="+", command=lambda: calcular("+"), 
          width=5, height=2, bg="lightblue").grid(row=0, column=0, padx=2)
tk.Button(frame_botones, text="-", command=lambda: calcular("-"), 
          width=5, height=2, bg="lightgreen").grid(row=0, column=1, padx=2)
tk.Button(frame_botones, text="√ó", command=lambda: calcular("*"), 
          width=5, height=2, bg="lightyellow").grid(row=0, column=2, padx=2)
tk.Button(frame_botones, text="√∑", command=lambda: calcular("/"), 
          width=5, height=2, bg="lightcoral").grid(row=0, column=3, padx=2)

# 6. Bot√≥n para limpiar
def limpiar():
    entrada1.delete(0, tk.END)
    entrada2.delete(0, tk.END)
    resultado_label.config(text="Resultado: --", fg="blue")
    print("Campos limpiados")

tk.Button(ventana, text="Limpiar", command=limpiar,
          bg="gray", fg="white").pack(pady=10)

# 7. Iniciar aplicaci√≥n
ventana.mainloop()
```

```python
"""
Ejemplo 2: Calculadora B√°sica
Diplomado en Mecatr√≥nica - UCAB
"""
import tkinter as tk

class CalculadoraBasica:
    def __init__(self, ventana):
        self.ventana = ventana
        self.ventana.title("Calculadora Mecatr√≥nica")
        self.ventana.geometry("350x450")
        self.ventana.resizable(False, False)
        
        # Variables para almacenar los n√∫meros y operaci√≥n
        self.numero1 = tk.StringVar()
        self.numero2 = tk.StringVar()
        self.resultado = tk.StringVar()
        self.resultado.set("Resultado aparecer√° aqu√≠")
        
        # Color de fondo moderno
        self.ventana.configure(bg='#2c3e50')
        
        self.crear_widgets()
    
    def crear_widgets(self):
        # T√≠tulo con estilo
        titulo = tk.Label(
            self.ventana,
            text="CALCULADORA B√ÅSICA",
            font=("Arial", 18, "bold"),
            bg='#2c3e50',
            fg='#ecf0f1'
        )
        titulo.pack(pady=15)
        
        # Frame para entrada de datos
        frame_entradas = tk.Frame(self.ventana, bg='#34495e', padx=10, pady=10)
        frame_entradas.pack(pady=10, padx=20, fill="x")
        
        # Entrada para primer n√∫mero
        tk.Label(
            frame_entradas,
            text="Primer n√∫mero:",
            font=("Arial", 11),
            bg='#34495e',
            fg='#ecf0f1'
        ).grid(row=0, column=0, sticky="w", pady=5)
        
        entrada1 = tk.Entry(
            frame_entradas,
            textvariable=self.numero1,
            font=("Arial", 12),
            width=15,
            justify="center"
        )
        entrada1.grid(row=0, column=1, pady=5, padx=10)
        
        # Entrada para segundo n√∫mero
        tk.Label(
            frame_entradas,
            text="Segundo n√∫mero:",
            font=("Arial", 11),
            bg='#34495e',
            fg='#ecf0f1'
        ).grid(row=1, column=0, sticky="w", pady=5)
        
        entrada2 = tk.Entry(
            frame_entradas,
            textvariable=self.numero2,
            font=("Arial", 12),
            width=15,
            justify="center"
        )
        entrada2.grid(row=1, column=1, pady=5, padx=10)
        
        # Frame para botones de operaciones
        frame_botones = tk.Frame(self.ventana, bg='#2c3e50', pady=20)
        frame_botones.pack()
        
        # Botones de operaciones
        operaciones = [
            ("‚ûï Sumar", self.sumar, "#27ae60"),
            ("‚ûñ Restar", self.restar, "#e74c3c"),
            ("‚úñ Multiplicar", self.multiplicar, "#3498db"),
            ("‚ûó Dividir", self.dividir, "#f39c12")
        ]
        
        for i, (texto, comando, color) in enumerate(operaciones):
            btn = tk.Button(
                frame_botones,
                text=texto,
                command=comando,
                bg=color,
                fg="white",
                font=("Arial", 11, "bold"),
                width=15,
                height=2,
                relief="raised",
                cursor="hand2"
            )
            btn.grid(row=i//2, column=i%2, padx=5, pady=5)
        
        # Frame para resultado
        frame_resultado = tk.Frame(self.ventana, bg='#34495e', padx=10, pady=10)
        frame_resultado.pack(pady=20, padx=20, fill="x")
        
        tk.Label(
            frame_resultado,
            text="RESULTADO:",
            font=("Arial", 12, "bold"),
            bg='#34495e',
            fg='#ecf0f1'
        ).pack(pady=(0, 10))
        
        self.etiqueta_resultado = tk.Label(
            frame_resultado,
            textvariable=self.resultado,
            font=("Arial", 14, "bold"),
            bg='white',
            fg='#2c3e50',
            width=25,
            height=2,
            relief="sunken"
        )
        self.etiqueta_resultado.pack()
        
        # Bot√≥n para limpiar
        tk.Button(
            self.ventana,
            text="üîÑ Limpiar",
            command=self.limpiar,
            bg="#95a5a6",
            fg="white",
            font=("Arial", 10),
            width=10
        ).pack(pady=10)
        
        # Informaci√≥n
        tk.Label(
            self.ventana,
            text="Ingresa dos n√∫meros y selecciona una operaci√≥n",
            font=("Arial", 9),
            bg='#2c3e50',
            fg='#bdc3c7'
        ).pack(pady=5)
    
    def obtener_numeros(self):
        """Obtiene y valida los n√∫meros ingresados"""
        try:
            num1 = float(self.numero1.get())
            num2 = float(self.numero2.get())
            return num1, num2
        except ValueError:
            self.resultado.set("ERROR: Ingresa n√∫meros v√°lidos")
            return None, None
    
    def sumar(self):
        num1, num2 = self.obtener_numeros()
        if num1 is not None:
            resultado = num1 + num2
            self.resultado.set(f"{num1} + {num2} = {resultado}")
            print(f"Operaci√≥n: {num1} + {num2} = {resultado}")
    
    def restar(self):
        num1, num2 = self.obtener_numeros()
        if num1 is not None:
            resultado = num1 - num2
            self.resultado.set(f"{num1} - {num2} = {resultado}")
            print(f"Operaci√≥n: {num1} - {num2} = {resultado}")
    
    def multiplicar(self):
        num1, num2 = self.obtener_numeros()
        if num1 is not None:
            resultado = num1 * num2
            self.resultado.set(f"{num1} √ó {num2} = {resultado}")
            print(f"Operaci√≥n: {num1} √ó {num2} = {resultado}")
    
    def dividir(self):
        num1, num2 = self.obtener_numeros()
        if num1 is not None:
            if num2 == 0:
                self.resultado.set("ERROR: No se puede dividir entre 0")
                print("Error: Divisi√≥n entre cero")
            else:
                resultado = num1 / num2
                self.resultado.set(f"{num1} √∑ {num2} = {resultado:.4f}")
                print(f"Operaci√≥n: {num1} √∑ {num2} = {resultado:.4f}")
    
    def limpiar(self):
        self.numero1.set("")
        self.numero2.set("")
        self.resultado.set("Resultado aparecer√° aqu√≠")
        print("Calculadora limpiada")

# Ejecutar la aplicaci√≥n
if __name__ == "__main__":
    root = tk.Tk()
    app = CalculadoraBasica(root)
    
    # Centrar la ventana en la pantalla
    root.update_idletasks()
    ancho = root.winfo_width()
    alto = root.winfo_height()
    x = (root.winfo_screenwidth() // 2) - (ancho // 2)
    y = (root.winfo_screenheight() // 2) - (alto // 2)
    root.geometry(f'{ancho}x{alto}+{x}+{y}')
    
    root.mainloop()
```

```python
import tkinter as tk

# -----------------------------
# Game constants
# -----------------------------
WIDTH = 800
HEIGHT = 600

PLAYER_SIZE = 30
ENEMY_SIZE = 30

MOVE_SPEED = 6
ENEMY_SPEED = 2

GRAVITY = 1.2
JUMP_FORCE = -18

LEVEL_WIDTH = 2000

MAX_LIVES = 3
INVINCIBILITY_TIME = 120

# -----------------------------
# Window
# -----------------------------
root = tk.Tk()
root.title("2D Game - Jupyter Friendly")

canvas = tk.Canvas(root, width=WIDTH, height=HEIGHT, bg="skyblue")
canvas.pack()

# -----------------------------
# HUD
# -----------------------------
lives = MAX_LIVES
hud = canvas.create_text(
    20, 20, anchor="nw",
    text=f"Lives: {lives}",
    font=("Arial", 16),
    fill="black"
)

# -----------------------------
# Platforms
# -----------------------------
platforms = [
    (0, HEIGHT - 60, LEVEL_WIDTH, HEIGHT),
    (200, 420, 400, 450),
    (500, 350, 700, 380),
    (900, 300, 1100, 330),
]

platform_ids = [
    canvas.create_rectangle(0, 0, 0, 0, fill="green")
    for _ in platforms
]

# -----------------------------
# Player
# -----------------------------
player_x = 100
player_y = 100
velocity_y = 0
on_ground = False

invincible_timer = 0
anim_timer = 0

player = canvas.create_rectangle(0, 0, 0, 0, fill="red")

# -----------------------------
# Enemies
# -----------------------------
enemies = [
    {
        "x": 300,
        "y": 0,
        "vx": ENEMY_SPEED,
        "vy": 0,
        "on_ground": False,
        "alive": True
    }
]


enemy_ids = [
    canvas.create_rectangle(0, 0, 0, 0, fill="purple")
    for _ in enemies
]

# -----------------------------
# Camera
# -----------------------------
camera_x = 0

# -----------------------------
# Keyboard
# -----------------------------
keys = set()

def key_down(event):
    keys.add(event.keysym)

def key_up(event):
    keys.discard(event.keysym)

root.bind("<KeyPress>", key_down)
root.bind("<KeyRelease>", key_up)

# -----------------------------
# Helpers
# -----------------------------
def rects_intersect(a, b):
    return not (
        a[2] <= b[0] or
        a[0] >= b[2] or
        a[3] <= b[1] or
        a[1] >= b[3]
    )

def take_damage():
    global lives, invincible_timer
    if invincible_timer > 0:
        return
    lives -= 1
    canvas.itemconfig(hud, text=f"Lives: {lives}")
    invincible_timer = INVINCIBILITY_TIME

# -----------------------------
# Game loop
# -----------------------------
def game_loop():
    global player_x, player_y, velocity_y, on_ground
    global camera_x, invincible_timer, anim_timer

    # -------------------------
    # PLAYER MOVE
    # -------------------------
    dx = 0
    moving = False

    if "Left" in keys:
        dx -= MOVE_SPEED
        moving = True
    if "Right" in keys:
        dx += MOVE_SPEED
        moving = True

    player_x += dx
    player_x = max(0, min(LEVEL_WIDTH - PLAYER_SIZE, player_x))

    player_rect = (player_x, player_y,
                   player_x + PLAYER_SIZE, player_y + PLAYER_SIZE)

    for plat in platforms:
        if rects_intersect(player_rect, plat):
            if dx > 0:
                player_x = plat[0] - PLAYER_SIZE
            elif dx < 0:
                player_x = plat[2]
            break

    # -------------------------
    # JUMP & GRAVITY
    # -------------------------
    if "space" in keys and on_ground:
        velocity_y = JUMP_FORCE
        on_ground = False

    velocity_y += GRAVITY
    player_y += velocity_y

    on_ground = False

    player_rect = (player_x, player_y,
                   player_x + PLAYER_SIZE, player_y + PLAYER_SIZE)

    for plat in platforms:
        if rects_intersect(player_rect, plat):
            if velocity_y > 0:
                player_y = plat[1] - PLAYER_SIZE
                velocity_y = 0
                on_ground = True
            elif velocity_y < 0:
                player_y = plat[3]
                velocity_y = 0
            break

    # -------------------------
    # PLAYER vs ENEMY
    # -------------------------
    for enemy in enemies:
        if not enemy["alive"]:
            continue

    enemy_rect = (
        enemy["x"], enemy["y"],
        enemy["x"] + ENEMY_SIZE,
        enemy["y"] + ENEMY_SIZE
    )

    if rects_intersect(player_rect, enemy_rect):
        if velocity_y > 0 and player_y + PLAYER_SIZE - 5 < enemy["y"]:
            enemy["alive"] = False
            velocity_y = JUMP_FORCE / 2
        else:
            take_damage()

    # -------------------------
    # ENEMY PHYSICS
    # -------------------------
    for enemy in enemies:
        if not enemy["alive"]:
            continue

        # Gravity
        enemy["vy"] += GRAVITY
        enemy["y"] += enemy["vy"]

        enemy_rect = (
            enemy["x"],
            enemy["y"],
            enemy["x"] + ENEMY_SIZE,
            enemy["y"] + ENEMY_SIZE
        )

        enemy["on_ground"] = False

        # Vertical collision
        for plat in platforms:
            if rects_intersect(enemy_rect, plat):
                if enemy["vy"] > 0:
                    enemy["y"] = plat[1] - ENEMY_SIZE
                    enemy["vy"] = 0
                    enemy["on_ground"] = True
                    break

        # Horizontal movement (only if grounded)
        if enemy["on_ground"]:
            enemy["x"] += enemy["vx"]

            front_x = enemy["x"] + (ENEMY_SIZE if enemy["vx"] > 0 else 0)
            foot_y = enemy["y"] + ENEMY_SIZE + 2

            # Edge detection
            supported = False
            for plat in platforms:
                if plat[0] <= front_x <= plat[2] and abs(plat[1] - foot_y) < 6:
                    supported = True
                    break

            if not supported:
                enemy["vx"] *= -1  # turn around


    # -------------------------
    # SIMPLE "ANIMATION"
    # -------------------------
    anim_timer += 1
    if invincible_timer > 0:
        invincible_timer -= 1
        canvas.itemconfig(
            player,
            fill="white" if anim_timer % 10 < 5 else "red"
        )
    elif not on_ground:
        canvas.itemconfig(player, fill="orange")  # jumping
    elif moving:
        canvas.itemconfig(player, fill="red" if anim_timer % 20 < 10 else "darkred")
    else:
        canvas.itemconfig(player, fill="red")

    # -------------------------
    # CAMERA
    # -------------------------
    camera_x = player_x - WIDTH // 2
    camera_x = max(0, min(LEVEL_WIDTH - WIDTH, camera_x))

    # -------------------------
    # DRAW PLAYER
    # -------------------------
    canvas.coords(
        player,
        player_x - camera_x, player_y,
        player_x - camera_x + PLAYER_SIZE,
        player_y + PLAYER_SIZE
    )

    # -------------------------
    # DRAW PLATFORMS
    # -------------------------
    for i, plat in enumerate(platforms):
        x1, y1, x2, y2 = plat
        canvas.coords(
            platform_ids[i],
            x1 - camera_x, y1,
            x2 - camera_x, y2
        )

    # -------------------------
    # DRAW ENEMIES
    # -------------------------
    for i, enemy in enumerate(enemies):
        if enemy["alive"]:
            canvas.coords(
                enemy_ids[i],
                enemy["x"] - camera_x, enemy["y"],
                enemy["x"] - camera_x + ENEMY_SIZE,
                enemy["y"] + ENEMY_SIZE
            )
        else:
            canvas.coords(enemy_ids[i], -100, -100, -50, -50)

    root.after(16, game_loop)

    canvas.itemconfig(
        enemy_ids[i],
        fill="purple" if enemy["on_ground"] else "pink"
    )

# -----------------------------
# Start
# -----------------------------
game_loop()
root.mainloop()
```
